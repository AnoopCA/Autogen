```python
import sys
import os
from autogen import ConversableAgent, GroupChat, GroupChatManager

# Add the parent directory to sys.path
sys.path.append(os.path.abspath(".."))

from llm_config import llm_config
```

## ðŸ§‘â€ðŸ³ Define Cake Baking Agents

```python
ingredient_agent = ConversableAgent(
    "IngredientAgent",
    llm_config = llm_config,
    system_message = "You are responsible for gathering ingredients for any cake requested. Respond in less than 25 words."
)

ingredient_agent.description = "Gather ingredients for the requested cake."


mixing_agent = ConversableAgent(
    name="MixingAgent",
    llm_config=llm_config,
    system_message="You mix all the ingredients provided into a batter. Respond in less than 25 words.",
)
mixing_agent.description = "Mix ingredients to prepare cake batter."

baking_agent = ConversableAgent(
    name="BakingAgent",
    llm_config=llm_config,
    system_message="You bake the batter at the appropriate temperature. Respond in less than 25 words.",
)
baking_agent.description = "Bake the cake batter in an oven."

decorating_agent = ConversableAgent(
    name="DecoratingAgent",
    llm_config=llm_config,
    system_message="You decorate the cake based on the requested flavor. Respond in less than 25 words.",
)
decorating_agent.description = "Decorate the baked cake with relevant toppings."

tasting_agent = ConversableAgent(
    name="TastingAgent",
    llm_config=llm_config,
    system_message="You taste the final cake and give feedback. Respond in less than 25 words.",
)
tasting_agent.description = "Taste the cake and provide the final review."

home_baker_agent = ConversableAgent(
    name="HomeBakerAgent",
    llm_config=llm_config,
    system_message="You are a home baker trying to bake cakes with the help of other agents. Respond in less than 25 words.",
)
```

```python
group_chat = GroupChat(
    agents = [
        ingredient_agent,
        mixing_agent,
        baking_agent,
        decorating_agent,
        tasting_agent
    ],
    messages = [],
    max_round = 6
)
```

```python
group_chat_manager = GroupChatManager(
    groupchat=group_chat,
    llm_config = llm_config,
)

chat_result = home_baker_agent.initiate_chat(
    group_chat_manager,
    message = "I want to bake a chocolate cake.",
    summary_method="reflection_with_llm",
)
```

```python
print(chat_result.summary)
```

```python
group_chat_with_intros = GroupChat(
    agents = [
        ingredient_agent,
        mixing_agent,
        baking_agent,
        decorating_agent,
        tasting_agent
    ],
    messages = [],
    max_round = 6,
    send_introductions=True,
)
```

```python
group_chat_manager_with_intros = GroupChatManager(
    groupchat=group_chat_with_intros,
    llm_config=llm_config,
)
```

```python
chat_result = home_baker_agent.initiate_chats(
    [
        {
            "recipient": group_chat_manager_with_intros,
            "message": "I want to bake a chocolate cake.",
        },
        {
            "recipient": group_chat_manager_with_intros,
            "message": "Now I want to bake a vanilla cake.",
        }
    ]
)
```
